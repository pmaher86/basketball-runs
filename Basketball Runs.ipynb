# Are runs in basketball more than just statistical fluctuation?
## Intro
A well known phenomenon among basketball fans is that of the "run", in which one team scores a substantial number of points while the other teams scores few or zero. Such runs are often attributed to things like momentum or in-game adjustments by one team and are a prominent feature of the subjective basketball-watching experience, but could they be just narrative over-fitting to statistical fluctuation? We can test this with statistics.
## Setup
Let's import the libraries and download the data, which is every play from the 2010-2011 regular season.

```python
import os, urllib, zipfile, sys, pandas as pd, numpy as np, matplotlib.pyplot as plt, re, IPython
%matplotlib inline
```

```python
if not os.path.isfile('playbyplay20102011reg20110416.txt'):
    print('Downloading data...')
    sys.stdout.flush()
    urllib.urlretrieve('http://www.basketballvalue.com/publicdata/playbyplay20102011reg20110416.zip', 'playbyplay20102011reg20110416.zip')
    print('Unzipping...')
    sys.stdout.flush()
    zfile = zipfile.ZipFile('playbyplay20102011reg20110416.zip')
    zfile.extract('playbyplay20102011reg20110416.txt')
    zfile.close()
    os.remove('playbyplay20102011reg20110416.zip')
    print('Data file ready!')
else:
    print('Already downloaded')
```

## Preprocessing
We'll do this in pandas. Let's take a look at the data

```python
df = pd.read_csv('playbyplay20102011reg20110416.txt',sep='\t',skiprows=1,names=['game','line','timeLeft','play'],index_col=['game','line'])
df.sort(inplace=True)
df.head()
```

And now lets extract which teams are playing

```python
df['t1']=pd.Series(df.index.get_level_values(0)).str[-6:-3].values
df['t2']=pd.Series(df.index.get_level_values(0)).str[-3:].values
df.head()
```

Here we'll extract information about the scores from the play column. Since many of the things we're interested in need to be analyzed on the game by game level, we'll often need to explicitly iterate over the games. At about 1200 games for the season, these calculations can thus be a little slow.

```python
def teamscore(x,team):
    rexp=re.search('\[(\S+) (\d+)\-(\d+)\]',x)
    if not (rexp is None) and rexp.groups()[0] == team:
        return int(rexp.groups()[1])
    else:
        return np.nan
    
df['t1score'], df['t2score'] = (np.nan, np.nan)
df['t1pts'], df['t2pts'] = (np.nan, np.nan)
i = 1
#this'll take a minute
games = df.index.levels[0].unique()
for game in games:
    IPython.display.clear_output(wait='True')
    print "Processing game {}...".format(i)
    sys.stdout.flush()
    i+=1
    gdf = df.loc[game,:] #don't use this for assignment (chained assignment is iffy)
    df.loc[(game,),'t1score'] = gdf['play'].map(lambda x: teamscore(x,gdf['t1'].irow(0))).values
    df.loc[(game,),'t2score'] = gdf['play'].map(lambda x: teamscore(x,gdf['t2'].irow(0))).values
    df.loc[(game,1),['t1score','t2score']] = 0 #games start at 0-0
    df.loc[(game,),['t1score','t2score']] = gdf[['t1score','t2score']].ffill().values
    df.loc[(game,),['t1pts','t2pts']] = gdf[['t1score','t2score']].diff().values
    df.loc[(game,1),['t1pts','t2pts']] = 0
IPython.display.clear_output(wait='True')
df.head(10)
```

## Calculating possessions
Because basketball teams take turns getting the ball, in order to calculate the probability of a run we'll need to be able to track possessions. We'll get information about possession from the play column. Backing out where possession changes hands from only this column is a little tricky; we'll need to explicity handle a number of cases where the team indicated in the play column is not the team with possession.

```python
def possession(x, team1, team2):
    rexp=re.search('\[([A-Z]{3})(.*)\]',x)
    if not (rexp is None):
        #cases where we get a team name but it doesn't tell us about possession
        if (('Foul' in x and not 'Turnover' in x) or 'Substitution' in x or 'Timeout' in x or 
            'Ejection' in x or ('Violation' in x and not 'Turnover' in x)):
            return np.nan
        elif rexp.groups()[0] == team1:
            return 1
        elif rexp.groups()[0] == team2:
            return 2
        else:
            return np.nan
    else:
        return np.nan

df['possession'],df['possnum']=(np.nan,np.nan)
i=1
for game in games:
    IPython.display.clear_output(wait='True')
    print "Processing game {}...".format(i)
    sys.stdout.flush()
    i+=1
    gdf = df.loc[game,:] #don't use this for assignment
    df.loc[(game,),'possession'] = gdf['play'].map(lambda x: possession(x,gdf['t1'].irow(0),gdf['t2'].irow(0))).values
    #we will backfill nan's because whoever has the ball now will do something with it in the future
    df.loc[(game,),'possession'] = gdf['possession'].bfill().values
    posschange = gdf['possession'].diff().abs()
    df.loc[(game,),'possnum']=posschange.cumsum().values
    df.loc[(game,1),'possnum'] = 0
    df.loc[(game,),'possnum'] = gdf['possnum'].ffill().values
IPython.display.clear_output(wait='True')
df.head(10)
```

Let's quickly make sure our possession counting is working. We'll take a look at the number of points scored in every possession for the entire season. If things are working most possessions shouldn't have more than three points scored.

```python
ptsPerPoss = pd.Series(0,index=np.arange(10))
for game in df.index.levels[0].unique():
    gdf = df.loc[game,:]
    t1stats = gdf.groupby('possnum')['t1pts'].sum()
    t2stats = gdf.groupby('possnum')['t2pts'].sum()
    ptsPerPoss = (ptsPerPoss + (t1stats+t2stats).value_counts()).fillna(ptsPerPoss)
print ptsPerPoss
ptsPerPoss.plot(kind='bar')
```

This looks feasible. Most possessions yield zero or two points, a significant number yield one or three points, and a very small number yield 4-6. These 4-6 point possessions are at least technically possible considering things like rebounded free throws, and-ones on three point shots, and technical fouls where possession is returned after the free throw. Let's take a look at a six-point possession just to double check.

```python
for game in df.index.levels[0].unique():
    gdf = df.loc[game,:]
    t1stats = gdf.groupby('possnum')['t1pts'].sum()
    if (t1stats==6).any():
        print game
        print t1stats[t1stats==6]
        break
```

## Calculating runs
Now let's parse the scoring data as a series of runs, defined as a series of plays where only one team scores.

```python
df['run']=np.nan
for game in games:
    gdf = df.loc[game,:]
    lastscore=pd.Series(np.nan,index=gdf.index)
    lastscore[gdf['t1pts']>0]=1
    lastscore[gdf['t2pts']>0]=2
    run = lastscore.ffill().diff().abs()
    run[1] = 0
    run.ffill(inplace=True)
    df.loc[game,'run'] = run.cumsum().values
df.head(20)
```

Now let's calculate how many points were scored in each run and count the run-score frequencies.

```python
ptsPerRun = pd.Series(0,index=np.arange(30))
for game in df.index.levels[0].unique():
    gdf = df.loc[game,:]
    t1stats = gdf.groupby('run')['t1pts'].sum()
    t2stats = gdf.groupby('run')['t2pts'].sum()
    ptsPerRun = (ptsPerRun + (t1stats+t2stats).value_counts()).fillna(ptsPerRun)
print ptsPerRun
ptsPerRun.plot(kind='bar')
```

Looks good. We've now extracted all the raw data we need, so let's try fitting this data with some statistics.
## Modeling scoring
First, let's look at all the final scores for all the games in league.

```python
gamescores = pd.DataFrame(index=df.index.levels[0].unique(),columns=df.columns)
for game in df.index.levels[0].unique():
    gamescores.ix[game] = df.loc[game,:].iloc[-1]
gamescores.head()
```

Let's compare this to a simple normal distribution with the mean and standard deviation we calculate from the data.

```python
from scipy.stats import norm
allscores = pd.concat([gamescores['t1score'],gamescores['t2score']]).astype('float')
allscores.hist(bins = np.arange(allscores.min()-0.5,allscores.max()+1))
x = np.linspace(allscores.min(),allscores.max())
dist = norm(loc=allscores.mean(),scale=allscores.std())
plt.plot(x, dist.pdf(x)*len(allscores), linewidth=3)
```

Looks like a pretty good fit! Now let's look at the distribution of the difference of team scores. We'll compare this to the probability distribution of the difference of two normal random variables, which is itself a normal distribution (see [here](http://mathworld.wolfram.com/NormalDifferenceDistribution.html)).

```python
scorediffs = gamescores['t1score']-gamescores['t2score']
scorediffs.hist(bins=np.arange(scorediffs.min()-0.5,scorediffs.max()+1))
x=np.linspace(scorediffs.min(),scorediffs.max())
dist=norm(loc=0,scale=allscores.std()*np.sqrt(2))
plt.plot(x,dist.pdf(x)*len(scorediffs),linewidth=3)
plt.xlim(scorediffs.min(),scorediffs.max())
```

Here we see significant deviation from the normal distribution, indicating that the scores of teams 1 and 2 are not independent variables. And one would hope not! For one thing, trivially, the rules of basketball forbid the difference between team scores from being zero at the end of a game. But also, teams clearly adjust their play tactics in response to the score difference as the game goes on, particularly if the game is a close one. This is likely why we see significantly more observed small margins of victory: teams that are winning by a large margin often pull their starters in order to rest them, knowing that the second-string players might give away some of the lead but not enough to lose. In fact, lets test score differece at the end of different quarters and see how it evolves:

```python
plt.subplots(2,2,figsize=(15,6))
for q,i in zip(["1st Quarter", "2nd Quarter", "3rd Quarter", "4th Quarter"],range(4)):
    qscores = df[df['play'].str.contains("End of " + q)]
    allqscores=pd.concat([qscores['t1score'],qscores['t2score']]).astype('float')
    qscorediffs = qscores['t1score']-qscores['t2score']
    plt.subplot(2,2,i+1)
    qscorediffs.hist(bins=np.arange(qscorediffs.min()-0.5,qscorediffs.max()+1))
    x=np.linspace(qscorediffs.min(),qscorediffs.max())
    dist=norm(loc=0,scale=allqscores.std()*np.sqrt(2))
    plt.plot(x,dist.pdf(x)*len(qscorediffs),linewidth=3)
    plt.xlim(qscorediffs.min()-0.5,qscorediffs.max()+0.5)
    plt.title(q)
```

Interestingly, the first quarter agrees quite well with the predictions of a model where the scoring of competing teams is statistically independent! By the end of the fourth quarter, though, we see significant disagreement with the predictions of independent scoring. A couple notes: the slight left skew of the plots is a real and expected effect--home court advantage. Also, the huge spike at zero difference at the end of the fourth quarter is clearly the result of teams in a close game "playing for overtime".
## Analyzing runs
Now let's look at our original question and construct a simple model to test whether the scoring runs observed in the NBA disagree with predictions of possessions being statistically independent. 
### Assumptions
First, we'll make the major simplification that there are only two results of a possession: some points, or zero. This is obviously brushing under the rug the possibility of different numbers of points being scored in each possession, but the tradeoff will be a simple closed form solution for our model's predictions. Additionally, 80% of scoring possessions are 2 points, making the average points per scoring posession very close to 2 with relatively small variance. We'll also make another major simplying assumption: that all possessions in all games have the same change of being scoring possessions. We'll discuss possible consequences of these assumptions later.
### The Model
In this model, a run consists of N scores by one team and N-1 non-scores followed by a score by the other team (this final score is what ends the run). Additionally, an arbitrary number of paired non-scoring possessions can occur in the middle of a run. The probability of a run of a given number of scoring possessions is thus:

$$
\begin{eqnarray}
P(N) &=& \sum_{k=0}^{\infty}P(S)^N P(0)^{N-1}P(S) P(0)^{2k}{N+k-1 \choose k} \\ 
P(N) &=& P(S)^{N+1} P(0)^{N-1} \sum_{k=0}^{\infty} P(0)^{2^k}{N+k-1 \choose k} \\
P(N) &=& P(S)^{N+1} P(0)^{N-1} (1-P(0)^2)^{-N}
\end{eqnarray}
$$

where $P(N)$ is the probability of a run of $N$ scoring plays, $P(S)$ is the probability of a possession being a score, and $P(0)$ is the probability of a possession resulting in no points ($P(S)+P(0)=1$). To obtain the nice closed form expression at the end we've made use of a nice binomial series identity.

To start out, let's get the parameters of our model. We'll need the proportion of scoring and non-scoring possessions, as well as the average number of points per scoring possession.

```python
avg = (ptsPerPoss[1:]*ptsPerPoss.index[1:]).sum()/(ptsPerPoss[1:]).sum()
var = (ptsPerPoss[1:]*(ptsPerPoss.index[1:]**2)).sum()/(ptsPerPoss[1:]).sum() - avg**2
print "average points per possession: {}".format(avg)
print "std dev of points per possession: {}".format(np.sqrt(var))
print "probability of no points in a possession: {}".format(ptsPerPoss[0]/ptsPerPoss.sum())
```

And now let's test the predictions of our model:

```python
avgPts = 2.11
def runprob(N,p_nopts = 0.493):
    p_pts = 1 - p_nopts
    return (p_pts**(N+1))*(p_nopts**(N-1))*(1-p_nopts**2)**(-N)
ptsPerRun.plot(kind='bar')
x=np.arange(1,12)
runprobs=runprob(x)*ptsPerRun.sum()
plt.plot(x * avgPts, runprobs, 'g-o', linewidth=3, markersize=8)
plt.yscale('log')
```

This looks shockingly accurate. Despite our very naive model, the predictions about the number and value of runs throughout an entire season of basketball are predicted tremendously well. It feels pretty safe to say that the runs observed in the NBA are consistent with the very strong assumption that the results of individual possessions are statistically independent. Before we go too far though, let's just plot the data a couple more ways: while it looks like there's minimal deviation from our model [aside from 1 point streaks, which A) are not captured by our 2.1-points-per-score model and B) are not that interesting to our analysis of long runs and "momentum"], we're plotting on a log-axis, which can hide significant differences. Let's plot on a few linear scales:

```python
plt.subplots(1, 3,figsize=(20,6))
plt.subplot(1,3,1)
ptsPerRun.plot(kind='bar')
plt.plot(2.1*x,runprobs,'g-o',linewidth=3,markersize=8)
plt.ylim([0, 5000])

plt.subplot(1,3,2)
ptsPerRun.plot(kind='bar')
plt.plot(2.1*x,runprobs,'g-o',linewidth=3,markersize=8)
plt.ylim([0, 500])

plt.subplot(1,3,3)
ptsPerRun.plot(kind='bar')
plt.plot(2.1*x,runprobs,'g-o',linewidth=3,markersize=8)
plt.ylim([0, 50])
```

The deviation still looks pretty minimal. I have to say, I find this result really surprising. Even as someone aware of the common statistical fallacies people apply to games and sports, watching a basketball game there is often a visceral sense of momentum, where one team scores with ease while the other can't buy a basket, and it often really **looks** like something more than random chance. Additionally, there are factors like lineup mismatches within a game that you might expect to give one team a significantly higher scoring probability for a few minutes. However real this concept of momentum seems, though, the data can be fully explained without it. That is **not** to say that things like momentum doesn't exist, just that they're not detectable as something causing deviation from our naive statistical model. For instance, if you thought that a period of momentum which gave one team an 8 point run should occur, say, once a game *beyond* statistical fluctuation, then 8-point runs would happen *twice* as often as our model predicts.
## Limitations of the model and conclusions
It's obvious that this model makes some questionable simplifications: not all scoring possessions yield 2.1 points (in fact, none do) and not all possessions have the same probability of resulting in points (this is why good teams consistently beat bad teams). However, in making both of these assumptions, we have **underestimated** the probability of a run occurring. Allowing larger variance in points per possession should give us a broader distribution of run score totals, making high-score runs more likely. Allowing one team's scoring probability to be higher (or lower) should also lead to more runs. (If this isn't obvious, consider that scoring disparity *requires* runs of more than one possession, so games with high scoring disparity should feature more high-point runs.) Since we are underestimating the probability of long runs, the fact that we don't see significant deviation remains striking. As a final point, one might think that the model "cheats" by normalizing to the total number of runs. Because our counting of runs contains all points scored for the season, however, this is no different than normalizing using the number of possessions--what we care about is the relative frequency of runs, and our results hold quite well across all observed runs.
